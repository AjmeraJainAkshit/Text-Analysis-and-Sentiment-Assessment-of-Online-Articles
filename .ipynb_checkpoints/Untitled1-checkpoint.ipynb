```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Load the input.xlsx file (assuming it contains a column named 'URL' with article URLs)
input_file = 'input.xlsx'
df = pd.read_excel(input_file)

# Create a directory to save the extracted articles
output_dir = 'extracted_articles.txt'
os.makedirs(output_dir, exist_ok=True)

# Iterate through each URL
for index, row in df.iterrows():
    url = row['URL']
    url_id = row['URL_ID']

    try:
        # Fetch the webpage content
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract article title and text (modify this based on the structure of the webpage)
        article_title = soup.find('h1').text.strip()
        article_text = soup.find('div', class_='td-post-content tagdiv-type').text.strip()

        # Save the extracted article to a text file
        output_file = os.path.join(output_dir, f'{url_id}.txt')
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(f"Title: {article_title}\n\n")
            f.write(article_text)

        print(f"Extracted article from {url} and saved to {output_file}")
    except Exception as e:
        print(f"Error extracting article from {url}: {str(e)}")

print("Extraction completed!")

```
